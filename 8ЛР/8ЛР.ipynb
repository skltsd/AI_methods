```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('./audit_risk.csv')
dataset.head()
```

```python
X = dataset[['Sector_score']].values
y = dataset['Audit_Risk'].values
print("Матрица признаков (фрагмент):")
print(X[:5])
print("Зависимая переменная (фрагмент):")
print(y[:5])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Зависимость целевой переменной от признака')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
