```python
data2 = np.genfromtxt("wine.data", delimiter=",")
print(data2.shape)
print(type(data2))
print(type(data2[0]))
print(type(data2[0][4]))
print(data2[:10])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

#    
alcohol = [] # 
malic_acid = [] #  
color_intensity = [] #  
hue = [] # 

for dot in data2:
    alcohol.append(dot[1])
    malic_acid.append(dot[2])
    color_intensity.append(dot[9])
    hue.append(dot[10])

plt.figure(1)
class1, = plt.plot(alcohol[:59], malic_acid[:59], 'ro', label='Class 1')
class2, = plt.plot(alcohol[59:130], malic_acid[59:130], 'g^', label='Class 2')
class3, = plt.plot(alcohol[130:], malic_acid[130:], 'bs', label='Class 3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('alcohol')
plt.ylabel('malic acid')
plt.title('Wine Dataset: alcohol vs malic acid')

plt.figure(2)
class1, = plt.plot(alcohol[:59], color_intensity[:59], 'ro', label='Class 1')
class2, = plt.plot(alcohol[59:130], color_intensity[59:130], 'g^', label='Class 2')
class3, = plt.plot(alcohol[130:], color_intensity[130:], 'bs', label='Class 3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('alcohol')
plt.ylabel('color intensity')
plt.title('Wine Dataset: alcohol vs color intensity')

plt.figure(3)
class1, = plt.plot(alcohol[:59], hue[:59], 'ro', label='Class 1')
class2, = plt.plot(alcohol[59:130], hue[59:130], 'g^', label='Class 2')
class3, = plt.plot(alcohol[130:], hue[130:], 'bs', label='Class 3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('alcohol')
plt.ylabel('hue')
plt.title('Wine Dataset: alcohol vs hue')

plt.show()
```
